```python
import numpy as np

#membuat array dengan numpy
mylist = [1,2,3]
myarray = np.array(mylist)

print(type(mylist))
print(type(myarray))
print(myarray.shape)
print(myarray)
```

```python
#access data pada numpy array
mylist_1 = [[1,2,3], [4,5,6]]
myarray_1 = np.array(mylist_1)

print(myarray_1)
print(myarray_1.shape)

print("Baris pertama:", myarray_1[0])
print("Baris kedua:", myarray_1[1])
print("Kolom pertama:", myarray_1[:,0])
print("Kolom kedua:", myarray_1[:,1])
print("Baris pertama kolom ketiga:", myarray_1[0][2])
```

```python
#operasi aritmatika pada numpy array
myarray1 = np.array([1,2,3])
myarray2 = np.array([4,5,6])

print("Penjumlahan array: ", myarray1 + myarray2)
print("Perkalian array: ", myarray1 * myarray2)
```

```python
array = np.array([3, 6, 9, 12])
division = array/3

print(division)
print (type(division))
```

```python
#kita membuat basic line plot
import matplotlib.pyplot as plt
import numpy as np

myarray = np.array([1,2,3])

plt.plot(myarray)
plt.title('pyplot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#membuat scatter plot
x = np.array([1,2,3])
y = np.array([4,5,6])

plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# SOAL 1

import numpy as np

arr = np.arange(1, 11)
rata_rata = np.mean(arr)
print("Rata-rata:", rata_rata)

maks = np.max(arr)
min = np.min(arr)
print("Maksimum:", maks)
print("Minimum:", min)

jumlah = np.sum(arr)
print("Jumlah:", jumlah)
```

```python
# SOAL 2

import numpy as np

array_2d = np.random.randint(0, 21, size=(3, 3))

print("Array awal:")
print(array_2d)

array_2d += 5

print("\nArray setelah ditambah 5:")
print(array_2d)

max_per_column = np.max(array_2d, axis=0)

print("\nElemen maksimum di setiap kolom:")
max_per_column
```

```python
# SOAL 3

import numpy as np

even_array = np.arange(2, 21, 2)

print("Elemen pertama hingga ke-5:", even_array[:5])

print("Elemen dengan indeks ganjil:", even_array[1::2])
```

```python
# SOAL 4

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = x**2

plt.plot(x, y)

plt.title('Grafik y = x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# SOAL 5

import pandas as pd

data = {'Name': ['Anna', 'Ben', 'Charlie', 'Diana', 'Edward'],
        'Age': [23, 21, 25, 22, 24],
        'Score': [85, 90, 88, 79, 95]}
df = pd.DataFrame(data)

print("Mahasiswa dengan skor di atas 85:")
print(df[df['Score'] > 85])

df['Passed'] = df['Score'] >= 80
print("\nDataFrame dengan kolom Passed:")
df
```
